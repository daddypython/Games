# Sales and profit over 30 months

```python
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(bedok["month"], bedok["customer"], color="#FF0000", label="Bedok")
ax.plot(kim["month"], kim["customer"], color="#0000FF", label="Kim Keat")
ax.plot(telok["month"], telok["customer"], color="#FFFF00", label="Telok Blangah")
ax.set_xlabel("Month: Jan 2019 to May 2021 - 0 is Jan 2019 and 30 is May 2021")
ax.set_ylabel("Number of cutomers")
plt.legend(loc='upper left')
plt.show()

style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(bedok["month"], bedok["sale"], color="#FF0000", label="Bedok")
ax.plot(kim["month"], kim["sale"], color="#0000FF", label="Kim Keat")
ax.plot(telok["month"], telok["sale"], color="#FFFF00", label="Telok Blangah")
ax.set_xlabel("Month: Jan 2019 to May 2021 - 0 is Jan 2019 and 30 is May 2021")
ax.set_ylabel("Sales")
plt.legend(loc='upper left')
plt.show()

style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(bedok["month"], bedok["income"], color="#FF0000", label="Bedok")
ax.plot(kim["month"], kim["income"], color="#0000FF", label="Kim Keat")
ax.plot(telok["month"], telok["income"], color="#FFFF00", label="Telok Blangah")
ax.set_xlabel("Month: Jan 2019 to May 2021 - 0 is Jan 2019 and 30 is May 2021")
ax.set_ylabel("Income before giving 7-11 56% of the share")
plt.legend(loc='upper left')
plt.show()


style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(bedok["month"], bedok["profit"], color="#FF0000", label="Bedok")
ax.plot(kim["month"], kim["profit"], color="#0000FF", label="Kim Keat")
ax.plot(telok["month"], telok["profit"], color="#FFFF00", label="Telok Blangah")
ax.set_xlabel("Month: Jan 2019 to May 2021 - 0 is Jan 2019 and 30 is May 2021")
ax.set_ylabel("Income after giving 7-11 56%")
plt.legend(loc='upper left')
plt.show()
```

# Sales and profit by year

```python
bedok_group = bedok.groupby("year")
telok_group = telok.groupby("year")
kim_group = kim.groupby("year")

bedok_2019 = bedok_group.get_group(2019)
bedok_2020 = bedok_group.get_group(2020)
bedok_2021 = bedok_group.get_group(2021)

telok_2019 = telok_group.get_group(2019)
telok_2020 = telok_group.get_group(2020)
telok_2021 = telok_group.get_group(2021)

kim_2019 = kim_group.get_group(2019)
kim_2020 = kim_group.get_group(2020)
kim_2021 = kim_group.get_group(2021)

style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(bedok_2019["month"], bedok_2019["profit"], color="#FF0000", label="2019")
ax.plot(bedok_2020["month"], bedok_2020["profit"], color="#0000FF", label="2020")
ax.plot(bedok_2021["month"], bedok_2021["profit"], color="#FFFF00", label="2021")
ax.set_ylabel("Bedok Profit over 3 years")
ax.set_xlabel("Month: Jan 2019 to May 2021 - 0 is Jan 2019 and 30 is May 2021")
plt.legend(loc='upper left')
ax.text(25, 10000, "Average Profit in 2021: $10,120", c="blue", alpha=0.5, fontsize=12, weight="bold", rotation=60)
ax.text(16, 8500, "Average Profit in 2020: $9,221", c="blue", alpha=0.5, fontsize=12, weight="bold")
ax.text(1, 9300, "Average Profit in 2019: $8,285", c="blue", alpha=0.5, fontsize=12, weight="bold")
plt.show()

style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(telok_2019["month"], telok_2019["profit"], color="#FF0000", label="2019")
ax.plot(telok_2020["month"], telok_2020["profit"], color="#0000FF", label="2020")
ax.plot(telok_2021["month"], telok_2021["profit"], color="#FFFF00", label="2021")
ax.set_ylabel("Telok Blangah Profit over 3 years")
ax.set_xlabel("Month: Jan 2019 to May 2021 - 0 is Jan 2019 and 30 is May 2021")
plt.legend(loc='upper left')
ax.text(25, 11000, "Average Profit in 2021: $11,623", c="blue", alpha=0.5, fontsize=12, weight="bold", rotation=50)
ax.text(14.7, 10500, "Average Profit in 2020: $11,720", c="blue", alpha=0.5, fontsize=12, weight="bold")
ax.text(1, 10100, "Average Profit in 2019: $9,250", c="blue", alpha=0.5, fontsize=12, weight="bold")

plt.show()

style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(kim_2019["month"], kim_2019["profit"], color="#FF0000", label="2019")
ax.plot(kim_2020["month"], kim_2020["profit"], color="#0000FF", label="2020")
ax.plot(kim_2021["month"], kim_2021["profit"], color="#FFFF00", label="2021")
ax.set_ylabel("Kim Keat Profit over 3 years")
ax.set_xlabel("Month: Jan 2019 to May 2021 - 0 is Jan 2019 and 30 is May 2021")
plt.legend(loc='upper left')
ax.text(25, 10500, "Average Profit in 2021: $10,710", c="blue", alpha=0.5, fontsize=12, weight="bold", rotation=70)
ax.text(14, 9750, "Average Profit in 2020: $10,932", c="blue", alpha=0.5, fontsize=12, weight="bold", rotation=60)
ax.text(1, 11750, "Average Profit in 2019: $10,895", c="blue", alpha=0.5, fontsize=12, weight="bold")


plt.show()
```
